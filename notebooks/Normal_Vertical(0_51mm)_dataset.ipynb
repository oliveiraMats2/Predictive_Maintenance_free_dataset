```python
import pandas as pd
import numpy as np
import os
```

```python
normal = '/content/normal'
failure = '/content/vertical-misalignment/0.51mm'
```

```python
normal_csv = []
for p, _, files in os.walk(os.path.abspath(normal)):
    for file in files:
        normal_csv.append(os.path.join(p, file))
normal_csv = sorted(normal_csv)
print(normal_csv)
```

```python
data = []
for i in range(len(normal_csv)):
  data.append(pd.read_csv(normal_csv[i]))
print(data)
```

```python
data = np.array(data)
print(data.shape)
```

```python
data = np.expand_dims(data, axis=-1)
print(data.shape)
```

```python
X_train = data[:34]
print(X_train.shape)
X_val = data[34:42]
print(X_val.shape)
X_test = data[42:]
print(X_test.shape)
```

```python
import h5py

with h5py.File('X_train_normal.h5', 'w') as hf:
    hf.create_dataset("X_train",  data=X_train)
```

```python
with h5py.File('X_val_normal.h5', 'w') as hf:
    hf.create_dataset("X_val",  data=X_val)
```

```python
with h5py.File('X_test_normal.h5', 'w') as hf:
    hf.create_dataset("X_test",  data=X_test)
```

```python
y_train = np.zeros((34))
y_val = np.zeros((8))
y_test = np.zeros((7))
print(y_train.shape, y_val.shape, y_test.shape)
```
