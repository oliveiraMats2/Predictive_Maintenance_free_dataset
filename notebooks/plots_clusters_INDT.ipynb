```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np

figure(figsize=(8, 6), dpi=120)

INDT_machine = 3 + 0.75
local_machine = 7 + 0.75

step = 1
end = 100

iterators = 8000

y_vector_indt_machine = iterators*np.arange(0, end, step)
y_vector_local_machine = iterators*np.arange(0, end, step)

x_vector_indt_machine = INDT_machine*np.arange(0, end, step)
x_vector_local_machine = local_machine*np.arange(0, end, step)

plt.ylim(0, max(y_vector_indt_machine))
plt.xlim(0, max(x_vector_indt_machine))

#plt.xticks(np.arange(0, 51, 5)) 
#plt.yticks(np.arange(0, 11, 1))

x_init_coord_local = 77.5
x_init_coord_INDT = 37.5

y_init_coord = 80000

plt.plot(x_vector_indt_machine, y_vector_indt_machine, color="green", label="Machine INDT")
plt.plot(x_vector_local_machine, y_vector_local_machine, color="blue", label="Local machine TPV")
plt.title(f"Comparao INDT x computador local com {y_init_coord} iteraes")

plt.legend(loc="upper left")
plt.xlabel("tempo (min)")
plt.ylabel("n iteraes")

plt.annotate("77 min", # this is the value which we want to label (text)
             (x_init_coord_local, y_init_coord), # x and y is the points location where we have to label
             textcoords="offset points",
             xytext=(0,10), # this for the distance between the points
             # and the text label
             ha='center',
             arrowprops=dict(arrowstyle="->", color='green'))

plt.annotate("37 min", # this is the value which we want to label (text)
             (x_init_coord_INDT, y_init_coord), # x and y is the points location where we have to label
             textcoords="offset points",
             xytext=(0,10), # this for the distance between the points
             # and the text label
             ha='center',
             arrowprops=dict(arrowstyle="->", color='green'))
```
